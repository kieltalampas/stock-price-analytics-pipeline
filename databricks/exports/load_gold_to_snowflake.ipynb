```python
# Databricks notebook source
df_dim = spark.read.table('stock_prices.stocks_dlt.dim_stock')
df_fact = spark.read.table('stock_prices.stocks_dlt.fact_stock_prices_gold')
```

```python
(
    df_dim
    .coalesce(1)
    .write
    .mode("overwrite")
    .option("header", "true")
    .csv("/Volumes/stock_prices/export/dim_stock")
)
```

```python
(
    df_fact
    .coalesce(1)
    .write
    .mode("overwrite")
    .option("header", "true")
    .csv("/Volumes/stock_prices/export/fact_stock_price")
)
```
