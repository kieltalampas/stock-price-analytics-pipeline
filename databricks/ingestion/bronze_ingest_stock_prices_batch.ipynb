```python
# Databricks notebook source
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
schema = StructType([
    StructField("Date", StringType(), True),
    StructField("Open", DoubleType(), True),
    StructField("High", DoubleType(), True),
    StructField("Low", DoubleType(), True),
    StructField("Close", DoubleType(), True),
    StructField("Volume", DoubleType(), True),
    StructField("Brand_Name", StringType(), True),
    StructField("Ticker", StringType(), True),
    StructField("Industry_Tag", StringType(), True),
    StructField("Country", StringType(), True),
    StructField("Dividends", DoubleType(), True),
    StructField("Stock_Splits", DoubleType(), True),
    StructField("Capital_Gains", DoubleType(), True)
])
```

```python
df = spark.read.option('header', True).schema(schema).csv('/Volumes/stock_prices/source/world_stock_prices')
df.printSchema()
```
